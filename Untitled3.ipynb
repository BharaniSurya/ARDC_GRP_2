```python
import pandas as pd
df = pd.read_csv('Hurricane_Harvey.csv', header= 0,
                        encoding= 'unicode_escape')
```

```python
df = pd.DataFrame(df)

data_types = df.dtypes

print(data_types)
```

```python
def convert_to_str(x):
    try:
        return str(x)
    except:
        return None
df['Tweet'] = df['Tweet'].apply(convert_to_str)
```

```python
import regex as re

df['Tweet'] = df['Tweet'].str.strip()  # Remove leading and trailing whitespace
df['Tweet'] = df['Tweet'].str.lower()  # Convert text to lowercase
df['Tweet'] = df['Tweet'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x))  # Remove special characters
```

```python
import emoji

# Sample tweet data with emojis and emoticons
data = {'tweets': [
    "I love Python! üòç",
    "Feeling sad today. üòû",
    "Party time! üéâ",
    "Having a great day :)",
]}

df = pd.DataFrame(data)

# Function to replace emojis and emoticons with descriptions
def map_emojis(text):
    # Replace emojis and emoticons with descriptions
    text_with_descriptions = emoji.demojize(text)
    return text_with_descriptions

# Apply the function to the 'tweets' column
df['cleaned_tweets'] = df['tweets'].apply(map_emojis)

# Display the original and cleaned tweets
print(df[['tweets', 'cleaned_tweets']])
```
