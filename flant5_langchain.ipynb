```python
import pandas as pd

df=pd.read_csv('sentiment-emotion-labelled_Dell_tweets.csv')
df
```

```python
df=df.head(2000)
```

```python
df=df[['Text','emotion']]
df
```

```python
import transformers
pipeline = transformers.pipeline(
    "summarization",
    model=model,
    tokenizer=tokenizer,
    #torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    #device_map=device_map,
    max_length=200,
    use_cache=True,

)
```

```python
from langchain import HuggingFacePipeline

#call the huggingface pipeline() object & pass the pipeline and model parameters
#temperature of 0, makes the model not to hallucinate much (make its own answers)
llm = HuggingFacePipeline(pipeline = pipeline, model_kwargs = {'temperature':0.5})

#Langchain contains PromptTemplate - allows to alter answers from the llm
#LLMChain - chains the prompttemplate and LLM together

from langchain import PromptTemplate, LLMChain
```

```python

template = """
You are an annotator tasked with selecting only one emotion from the following based on the emotion present in the given query, 
give answer in one word: anger, anticipation, disgust, fear, joy, optimism, sadness, surprise.
Question: {query}
Answer:"""
prompt = PromptTemplate(template=template, input_variables=["query"])

#chain
llm_chain = LLMChain(prompt=prompt, llm=llm)
```

```python
c=[]
for i in df['Text']:
    query = i
    a=llm_chain.run(query)
    w = a.split()
    c.append(w)
```

```python
len(c)
```

```python
d=[]
for i in range(2000):
    d.append(c[i][0])
```

```python
d
```

```python
d = [s[:-1] for s in d]

# Print the updated list
print(d)
```

```python
len(d)
```

```python
df1=df
df1
```

```python
df1['emotion_hat']= d
df1
```

```python
df1=df1[['emotion','emotion_hat']]
df1
```

```python
df1.to_csv("flanT5.csv")
```

```python
a=list(df1['emotion'])
b=list(df1['emotion_hat'])
```

```python
d=[]
for i in range(len(a)):
    if type(a[i])== str and type(b[i])== str:
        word_set1 = set(a[i])
        word_set2 = set(b[i])

        # Calculate Jaccard Similarity
        intersection = len(word_set1.intersection(word_set2))
        union = len(word_set1.union(word_set2))

        jaccard_similarity = intersection / union
        d.append(jaccard_similarity)
    else:
        d.append(0)
```

```python
df1["Similarity"]=d
df1
```

```python
sum1 = df1['Similarity'].sum()
print("Emotion score:",sum1)
```

```python
correct_predictions = (df1['emotion'] == df1['emotion_hat']).sum()
total_predictions = len(df1)
accuracy = correct_predictions / total_predictions

print("Accuracy:", accuracy)
```
