```python
import pandas as pd
df = pd.read_csv('TS_Harvey_Tweets.csv')
```

```python
df = pd.DataFrame(df)

data_types = df.dtypes

print(data_types)
```

```python
def convert_to_str(x):
    try:
        return str(x)
    except:
        return None
df['Tweet'] = df['Tweet'].apply(convert_to_str)
```

```python
import pandas as pd
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import TweetTokenizer
from nltk.stem import WordNetLemmatizer




#  Lowercasing
df['Tweet'] = df['Tweet'].str.lower()

#  Removal of Special Characters and Punctuation
df['Tweet'] = df['Tweet'].apply(lambda x: re.sub(r'[^\w\s]', '', x))


#  Handling Mentions  
df['Tweet'] = df['Tweet'].apply(lambda x: re.sub(r'@\w+', '', x))

# Removing hastags
df['Tweet'] = df['Tweet'].apply(lambda x: re.sub(r'#\w+', '', x))

#  URL Removal 
df['Tweet'] = df['Tweet'].apply(lambda x: re.sub(r'http\S+', '', x))




print(df)
```

```python
#  Tokenization using TweetTokenizer
tokenizer = TweetTokenizer()
df['Tweet'] = df['Tweet'].apply(tokenizer.tokenize)

# Lemmatization using NLTK
nltk.download("wordnet")
lemmatizer = WordNetLemmatizer()
df['Tweet'] = df['Tweet'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])


for index, row in df.iterrows():
    df.at[index, 'Tweet'] = ' '.join(row['Tweet'])

```
